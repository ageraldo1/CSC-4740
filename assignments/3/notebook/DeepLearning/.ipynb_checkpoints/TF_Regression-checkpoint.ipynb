#### Features Engineering
_____

```python
master = master.drop('id', axis=1)
```

```python
master['date'] = pd.to_datetime(master['date'])
```

```python
master['year'] = master['date'].apply(lambda date: date.year)
master['month'] = master['date'].apply(lambda date: date.month)
```

```python
sns.boxplot(x='month', y='price', data=master);
```

```python
master.groupby(by='month').mean()['price'].plot();
```

```python
master.groupby(by='year').mean()['price'].plot();
```

```python
master = master.drop('date', axis=1)
```

```python
master = master.drop('zipcode', axis=1)
```

#### Split Train and Test Data
________

```python
X = master.drop('price', axis=1).values    # NumPy array (TF requirement)
y = master['price'].values                 # NumPy array (TF requirement)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
print(f'X_train shape : {X_train.shape} - Y_train shape : {y_train.shape}')
print(f'X_test shape : {X_test.shape} - Y_test shape : {y_test.shape}')
```

#### Scale & Normalize Training Features
____

```python
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

#### Create Model
____

```python
model = Sequential()

model.add(Dense(units=19, activation='relu'))
model.add(Dense(units=19, activation='relu'))
model.add(Dense(units=19, activation='relu'))
model.add(Dense(units=19, activation='relu'))

#output layer
model.add(Dense(units=1))    

model.compile(optimizer='adam', loss='mse')   # regression
history = model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test), batch_size=128, epochs=400)
```

```python
losses = pd.DataFrame(history.history)
losses.plot();
```

#### Predictions
_____

```python
predictions = model.predict(X_test)
```

#### Evaluation
____

```python
print(f'Absolute Mean Error : {metrics.mean_absolute_error(y_test, predictions)}')
print(f'Mean Squared Error : {metrics.mean_squared_error(y_test, predictions)}')
print(f'Root Mean Squared Error : {np.sqrt(metrics.mean_squared_error(y_test, predictions))}')
```

```python
model.evaluate(X_test, y_test, verbose=0)
```

```python
metrics.explained_variance_score(y_test, predictions)
```

```python
plt.figure(figsize=(12,6))
plt.scatter(y_test, predictions);
plt.plot(y_test, y_test, 'r');   #perfect prediction
```
