#### Dependencies
_____

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

from tensorflow.keras.utils  import plot_model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.models import load_model
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

from mlxtend.plotting import plot_confusion_matrix

%matplotlib inline

sns.set_style("whitegrid")
```

#### Load Data
_____

```python
columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'label']
master = pd.read_csv('../resources/processed.cleveland.data', names=columns)
```

#### Data Cleanup
_____

```python
cond = master['ca'] == '?'
for row in master[cond]['ca'].index:
    master.iloc[row, 11] = master['ca'].mode().values
    
cond = master['thal'] == '?'
for row in master[cond]['thal'].index:
    master.iloc[row, 12] = master['thal'].mode().values
    
master['ca'] = master['ca'].astype('float64')
master['thal'] = master['thal'].astype('float64')
```

```python
master['label'] = master['label'].apply(lambda x: 0 if x == 0 else 1)
```

```python
def label_cp(x):
    if (x == 1):
        return 'typical-angina'
    
    elif (x == 2):
        return 'atypical-angina'
    
    elif (x == 3):
        return 'non-anginal-pain'
    
    else:
        return 'asymptomatic'
    
def label_restecg(x):
    if (x == 0):
        return 'normal'
    elif (x == 1):
        return 'wave-abnormality'
    else:
        return 'ventricular-hypertrophy'
    
def label_slope(x):
    if (x == 1):
        return 'upsloping'
    
    elif (x == 2):
        return 'flat'
    
    else:
        return 'downsloping'
    
def label_thal(x):
    if (x == 3):
        return 'normal'
    
    elif (x == 6):
        return 'fixed-defect'
    
    else:
        return 'reversable-defect'
    
def label_ca(x):
    if (x == 0):
        return 'zero'
    
    elif(x == 1):
        return 'one'
    
    elif(x == 2):
        return 'two'
    
    else:
        return 'three'        
        
master['sex'] = master['sex'].apply(lambda x: 'Male' if x == 1 else 'Female')
master['cp'] = master['cp'].apply(label_cp)
master['fbs'] = master['fbs'].apply(lambda x: 'False' if x == 0 else 'True')
master['restecg'] = master['restecg'].apply(label_restecg)
master['exang'] = master['exang'].apply(lambda x: 'No' if x == 0 else 'Yes')
master['slope'] = master['slope'].apply(label_slope)
master['thal'] = master['thal'].apply(label_thal)
master['ca'] = master['ca'].apply(label_ca)

data_d = master.copy(deep=True)
data_d['label'] = data_d['label'].apply(lambda x: 'Healthy' if x == 0 else 'Sick')
```

#### Encode Categorial Features
_____

```python
columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
dummies = pd.get_dummies(master[columns], drop_first=True)

master = master.drop(columns, axis=1)
master = pd.concat([master, dummies], axis=1)
```

#### Split Train and Test Data
______

```python
X = master.drop('label', axis=1).values    # NumPy array (TF requirement)
y = master['label'].values                 # NumPy array (TF requirement)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

print(f'X_train shape : {X_train.shape} - Y_train shape : {y_train.shape}')
print(f'X_test shape : {X_test.shape} - Y_test shape : {y_test.shape}')
```

#### Scale & Normalize Training Features
____

```python
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

#### Create Model
_____

##### Overfitting scenario (higher number of epochs)


```python
model = Sequential()

model.add(Dense(units=20, activation='relu'))
model.add(Dense(units=15, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))    

#model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.compile(optimizer='adam', loss='binary_crossentropy')
history = model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test))
```

```python
pd.DataFrame(history.history).plot();

plt.tight_layout();
```

+ Typical overfitting scenario:
    + the loss of validation set and the training set starts decreasing together (good signal)
    + after a certain amount of time (epoch around 50), the losses for the validation set starts to increase exponentially (validation loss getting worst and worst after 50 epochs)
    + when we have a scenario where the loss on training sets continue to decrease but the loss on validation set starts to increase, it indicate an overfitting issue.
    
    


#### Early Stop Approach
_____

```python
model = Sequential()

model.add(Dense(units=20, activation='relu'))
model.add(Dense(units=15, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))    

model.compile(optimizer='adam', loss='binary_crossentropy')
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=0, patience=10)
```

```python
history = model.fit(x=X_train, y=y_train, epochs=600, verbose=1, validation_data=(X_test,y_test), callbacks=[early_stop])
```

```python
pd.DataFrame(history.history).plot();

plt.tight_layout();
```

+ By monitoring the minimum value of val_loss, before it starts to increase again, we force the stop of the model training (patience = 10).
+ EarlyStop callback allow us to stop the model training before the validation loss starts to increase again.

#### Adding Dropout Layers
_____

+ Dropout layers helps also the model performance by turning off a set of random neurons

```python
model = Sequential()

model.add(Dense(units=20, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=15, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=1, activation='sigmoid'))    

model.compile(optimizer='adam', loss='binary_crossentropy')
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=0, patience=25)
```

```python
history = model.fit(x=X_train, y=y_train, epochs=600, verbose=1, validation_data=(X_test,y_test), callbacks=[early_stop])
```

```python
pd.DataFrame(history.history).plot();

plt.tight_layout();
```

#### Predictions
____

```python
predictions = model.predict_classes(X_test)
```

##### Evaluation
____

```python
print(classification_report(y_test, predictions))
```

##### Confusion Matrix
____

```python
fig, ax = plot_confusion_matrix(conf_mat=confusion_matrix(y_test, predictions),
                                show_absolute=True,
                                show_normed=True,
                                colorbar=True, class_names=['Healthy','Sick'],)
plt.show()
```
